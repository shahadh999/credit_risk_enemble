## Read the CSV and Perform Basic Data Cleaning

```python
# Load the data
file_path = Path('Resources/LoanStats_2019Q1.csv')
df = pd.read_csv(file_path)

# Preview the data
df.head()
```

## Split the Data into Training and Testing

```python
# Create our features
X = df.drop(columns=["loan_status"], axis=1)

# Create our target
y = df['loan_status']
y
```

```python
X.head()
```

```python
# Check the balance of our target values
df[["loan_status"]].value_counts()
```

```python
X.columns
```

```python
X_bin = pd.get_dummies(X, columns=["home_ownership", "verification_status", "issue_d", "pymnt_plan", "initial_list_status", "next_pymnt_d", "application_type", "hardship_flag", "debt_settlement_flag"])

X_bin.head()
```

```python
# Split the X and y into X_train, X_test, y_train, y_test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_bin,
                                                    y,
                                                    random_state=1)
```

## Data Pre-Processing

Scale the training and testing data using the `StandardScaler` from `sklearn`. Remember that when scaling the data, you only scale the features data (`X_train` and `X_testing`).

```python
# Create the StandardScaler instance
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
# When fitting scaling functions, only train on the training dataset
X_scaler = scaler.fit(X_train)
```

```python
# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)

X_test_scaled = X_scaler.transform(X_test)
```

## Ensemble Learners

### Balanced Random Forest Classifier

```python
# Resample the training data with the BalancedRandomForestClassifier
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=500,
                                  random_state=42)


rf_model.fit(X_train_scaled, y_train)
```

```python
# Calculated the balanced accuracy score
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

rf_pred = rf_model.predict(X_test_scaled)

rf_acc_score = accuracy_score(y_test, rf_pred)

print(f"The accuracy score for the Balanced Random Forest Classifier is {rf_acc_score}.")
```

```python
# Display the confusion matrix
rf_cm = confusion_matrix(y_test, rf_pred)

rf_cm_df = pd.DataFrame(rf_cm,
                        index=["Actual False", "Actual True"],
                        columns=["Predicted False", "Predicted True"])

rf_cm_df
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced

print(classification_report_imbalanced(y_test, rf_pred))
```

```python
# List the features sorted in descending order by feature importance
rf_importances = rf_model.feature_importances_

sorted(zip(rf_model.feature_importances_, X.columns), reverse=True)
```

### Easy Ensemble Classifier

```python
# Train the Classifier
from imblearn.ensemble import EasyEnsembleClassifier

easy_ens = EasyEnsembleClassifier(n_estimators=10,
                                  random_state=1)

easy_ens.fit(X_train_scaled, y_train)
```

```python
# Calculated the balanced accuracy score
easy_ens_pred = easy_ens.predict(X_test_scaled)

from sklearn.metrics import balanced_accuracy_score
easy_ens_bal_acc_score = balanced_accuracy_score(y_test, easy_ens_pred)

print(f"The balanced accuracy score for the Easy Ensemble Classifier is {easy_ens_bal_acc_score}")
```

```python
# Display the confusion matrix
easy_ens_cm = confusion_matrix(y_test, easy_ens_pred)

easy_ens_cm_df = pd.DataFrame(easy_ens_cm,
                              index=["Actual False", "Actual True"],
                              columns=["Predicted False", "Predicted True"])

easy_ens_cm_df
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y_test, easy_ens_pred))
```
